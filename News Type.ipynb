```python
import pandas as pd
```

```python
News_Input = pd.read_csv("C:/Hackathon/fake.csv", header = 0, encoding = "latin_1", sep = ",", low_memory = False)
```

```python
News_Input.head()
```

```python
News_Input["type"].value_counts()
```

```python
News_Type = pd.DataFrame({'count' : News_Input.groupby(["type"]).size()}).reset_index()
News_Type
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(len(News_Type["count"]))
y =  News_Type["count"]
xl = News_Type["type"]
##
fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(x,y)
ax.set_xticks(x)
ax.set_xticklabels(xl)
plt.show()
```
